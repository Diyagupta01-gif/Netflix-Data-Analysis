```python
import pandas as pd

# Load CSV
data = pd.read_csv(r"C:\Users\dines\Downloads\archive\mini python project\netflix_titles.csv")

# Check first 5 rows
data.head()
```

```python
# Check number of rows & columns
print(data.shape)

# Check column names
print(data.columns)

# Count Movies vs TV Shows
print(data['type'].value_counts())

# Check missing values
print(data.isnull().sum())
```

```python
# Remove duplicates (if any)
data = data.drop_duplicates()

# Fill missing values in 'country' column
data['country'] = data['country'].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='type', data=data)
plt.title('Movies vs TV Shows on Netflix')
plt.show()
```

```python
plt.figure(figsize=(12,5))       # Set chart size
sns.countplot(x='release_year', data=data)
plt.xticks(rotation=90)           # Rotate year labels for readability
plt.title('Movies/TV Shows Released Each Year')
plt.show()
```

```python
# Split 'listed_in' column by comma and create separate rows
all_genres = data['listed_in'].str.split(',').explode()

# Plot horizontal bar chart
plt.figure(figsize=(10,8))
sns.countplot(y=all_genres, order=all_genres.value_counts().index)
plt.title('Most Popular Genres on Netflix')
plt.show()
```
